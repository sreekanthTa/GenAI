Stemming

````python
# Stemming: Reduces words to its stems
#   ex: going -> go, goes -> go, gone -> go
# ex:  eating-> eat, eaten -> eat, Having variaty of eat not have much effect, since
#  They have same meaning as STEM or as a ROOT or as a BASE
# To reduc eno of tokens as inputs, we use stemming and lemmatization. Stemming is a rule based process, while lemmatization is a dictionary based process. Stemming is faster than lemmatization, but lemmatization is more accurate than stemming.
words = ["going", "goes", "gone", "eating", "eaten", "having", "happily", "happiness", "happier", "happiest","history","historical", "historically", "historian", "historians"]

     
````

1.Porter Stemmer

````python

# Porter Stemme
from nltk.stem  import PorterStemmer
porter_stemmer = PorterStemmer()

for word in words:
    print(word +" ------> "+porter_stemmer.stem(word))
       
# cons: For some of the words, it might not give correct STEM word, like histori
# happily ------> happili
# history ------> histori
````

2.RegexStemmer Class


````python
# Useful for applying stemming based on RegExp mat
# It removes the suffixes based on the regular expression pattern provided. 
# It is faster than Porter Stemmer, but it might not give correct STEM word for some of the words,
#  like histori
from nltk.stem import RegexpStemmer
regexp_stemmer = RegexpStemmer('ing$|s$|e$|ly$|i$|al$|ical$|ian$|ians$', min=4)
for word in words:
    print(word +" ------> "+regexp_stemmer.stem(word))

# cons: For some of the words, it might not give correct STEM word, like histori
# happily ------> happi
# history ------> histor
````

3. SnowBall Stemmer


````python
# Snowball Stemmer
# It is an improvement over Porter Stemmer. It is faster than Porter Stemmer, but
# it might not give correct STEM word for some of the words, like histori

from nltk.stem import SnowballStemmer
snowball_stemmer = SnowballStemmer('english')
for word in words:
    print(word +" ------> "+snowball_stemmer.stem(word))
````

3. LAMITIZATION

1.Wordnet Lamitizer

````python
# SWordnet Lamitizer: it gives ROOT word for all the words not STEM word
# pos : n= noun, v:verb, av:adverb

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

for word in words:
    print(word +" ------> "+lemmatizer.lemmatize(word, pos="v"))
````

````python
lamitizer = WordNetLemmatizer()
for i in range(len(sentences)):
    sentences[i] = sentences[i].lower() # convert to lower case
    words = word_tokenize(sentences[i])
    filtered_words = [word for word in words if word not in stopwords.words('english')]
    stemmed_words = [lamitizer.lemmatize(word, pos="v") for word in filtered_words]
    sentences[i] = " ".join(stemmed_words) 
    # print(stemmed_words)
print(sentences)
````

One-Hot Encoding Example

## Why One-Hot Encoding?

**Problem:** Machine learning models need numbers, not text categories.

**Solution:** Convert each category into binary (0 or 1) values.

**Example:** If we have 3 fruits, we create 3 new columns:
- Apple → [1, 0, 0]
- Banana → [0, 1, 0]
- Orange → [0, 0, 1]

This way, the model can understand and process categorical data!

````python
# Simple One-Hot Encoding Example
import pandas as pd

# Simple data: Just one category (Fruits)
fruits = ['Apple', 'Banana', 'Orange', 'Apple', 'Banana']

df = pd.DataFrame({'Fruit': fruits})
print("Original Data:")
print(df)
print("\n")

# Convert to one-hot encoding
one_hot = pd.get_dummies(df['Fruit'])
print("One-Hot Encoded:")
print(one_hot)
print("\nNotice: Each fruit is now a column with 1 or 0 values!")
````

````python
# One-Hot Encoding Example
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# Example data: Colors and Sizes
data = {
    'Color': ['Red', 'Blue', 'Green', 'Red', 'Blue', 'Green'],
    'Size': ['Small', 'Large', 'Medium', 'Medium', 'Small', 'Large']
}

df = pd.DataFrame(data)
print("Original Data:")
print(df)
print("\n")

# Method 1: Using pandas get_dummies
print("Method 1: Using pandas get_dummies")
one_hot_pandas = pd.get_dummies(df, columns=['Color', 'Size'])
print(one_hot_pandas)
print("\n")

# Method 2: Using sklearn OneHotEncoder
print("Method 2: Using sklearn OneHotEncoder")
encoder = OneHotEncoder(sparse_output=False)
one_hot_sklearn = encoder.fit_transform(df)
feature_names = encoder.get_feature_names_out(['Color', 'Size'])
one_hot_df = pd.DataFrame(one_hot_sklearn, columns=feature_names)
print(one_hot_df)
````

## Bag of Words

**What is Bag of Words?**

Bag of Words (BoW) converts text into numerical features by counting word frequency. It treats each document as a collection of words without considering their order.

**How it works:**
1. Create a vocabulary of all unique words from all documents
2. For each document, count how many times each word appears
3. Create a vector showing word counts

**Example:**
- Document 1: "cat likes milk"
- Document 2: "dog likes meat"

Vocabulary: [cat, dog, likes, meat, milk]

- Document 1 → [1, 0, 1, 0, 1]  (cat=1, dog=0, likes=1, meat=0, milk=1)
- Document 2 → [0, 1, 1, 1, 0]  (cat=0, dog=1, likes=1, meat=1, milk=0)

````python
# Simple Bag of Words Example
from sklearn.feature_extraction.text import CountVectorizer

# Sample documents
documents = [
    "cat likes milk",
    "dog likes meat",
    "cat likes meat",
    "dog likes milk"
]

# Create CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform
bow_matrix = vectorizer.fit_transform(documents)

# Get feature names (vocabulary)
feature_names = vectorizer.get_feature_names_out()

print("Vocabulary:", feature_names)
print("\nBag of Words Matrix:")
print(bow_matrix.toarray())
print("\nInterpretation:")
for i, doc in enumerate(documents):
    print(f"Document {i+1} '{doc}' → {bow_matrix.toarray()[i]}")
````

## TF-IDF (Term Frequency-Inverse Document Frequency)

**What is TF-IDF?**

TF-IDF is an improvement over Bag of Words. It not only counts word frequency but also considers how important a word is across all documents.

**Formula:**
- **TF (Term Frequency):** How often a word appears in a document
- **IDF (Inverse Document Frequency):** How rare/unique a word is across all documents
- **TF-IDF = TF × IDF**

**Why it's better than Bag of Words:**
- Common words (like "the", "is") get lower scores
- Unique/important words get higher scores
- Better for finding relevant documents

**Disadvantages:**
- Doesn't consider word order or context
- Ignores semantic meaning of words
- Can still give high scores to irrelevant frequent words if they're rare
- Not effective for very short documents
- Requires good preprocessing (stopword removal, stemming)

````python
# TF-IDF Example
from sklearn.feature_extraction.text import TfidfVectorizer

# Same documents
documents = [
    "cat likes milk",
    "dog likes meat",
    "cat likes meat",
    "dog likes milk",
    "cat is cute",
    "dog is friendly"
]

# Create TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

# Get feature names (vocabulary)
feature_names = tfidf_vectorizer.get_feature_names_out()

print("Vocabulary:", feature_names)
print("\nTF-IDF Matrix:")
print(tfidf_matrix.toarray())
print("\n--- Comparison ---")
print("Bag of Words: All words counted equally")
print("TF-IDF: Common words ('is', 'likes') get lower scores")
print("        Unique words ('cat', 'dog', 'meat') get higher scores")
print("\nNotice the decimal values - these are TF-IDF scores (0 to 1)")
````

## Difference Between Bag of Words vs One-Hot Encoding

| Feature | Bag of Words (BoW) | One-Hot Encoding |
|---------|-------------------|------------------|
| **Purpose** | Encodes text documents | Encodes categorical variables |
| **Input** | Text/Documents (multiple words) | Single categorical value per row |
| **Output** | Word frequency counts | Binary (0 or 1) per category |
| **Use Case** | Text classification, sentiment analysis, search | Categorical data (color, size, brand) |
| **Example Input** | "cat likes milk dog" | Color = "Red" |
| **Example Output** | [1, 1, 1, 1] (counts) | [1, 0, 0] (Red, Blue, Green) |
| **Vocabulary** | All unique words in documents | All unique categories |
| **Dimensions** | Number of unique words | Number of unique categories |
| **Information Loss** | No word order information | Lost relationship info (if exists) |

**Key Difference:**
- **BoW:** Used for TEXT - counts how many times each word appears
- **One-Hot:** Used for CATEGORIES - creates binary columns for each category value

**Example:**
```
Text: "The cat and dog"
BoW: [1, 1, 1, 1]  (counts: the=1, cat=1, and=1, dog=1)

Category: Color = "Red"
One-Hot: [1, 0, 0]  (Red=1, Blue=0, Green=0)
```

````python
# Comparison Example: BoW vs One-Hot Encoding
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

print("=" * 60)
print("BAG OF WORDS (BoW)")
print("=" * 60)

# Text data - multiple words per sample
texts = ["cat likes milk", "dog likes meat"]

vectorizer = CountVectorizer()
bow_result = vectorizer.fit_transform(texts).toarray()

print("Input (Text):", texts)
print("Vocabulary:", vectorizer.get_feature_names_out())
print("Output (BoW):\n", bow_result)
print("Shape:", bow_result.shape, "- (samples, word_features)\n")

print("=" * 60)
print("ONE-HOT ENCODING")
print("=" * 60)

# Categorical data - single category per sample
categories = pd.DataFrame({
    'Animal': ['cat', 'dog', 'cat', 'dog']
})

one_hot_result = pd.get_dummies(categories['Animal']).values

print("Input (Category):", categories['Animal'].values)
print("Output (One-Hot):\n", one_hot_result)
print("Shape:", one_hot_result.shape, "- (samples, category_features)\n")

print("=" * 60)
print("KEY DIFFERENCE:")
print("=" * 60)
print("BoW:      Works with TEXT (documents with many words)")
print("One-Hot:  Works with CATEGORIES (single categorical values)")
print("BoW:      Output = Frequency counts (0, 1, 2, 3...)")
print("One-Hot:  Output = Binary flags (0 or 1 only)")
````
