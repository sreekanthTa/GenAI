# Text Encoding Techniques

This notebook demonstrates common text encoding techniques used in natural language processing:

1. **One-Hot Encoding**: Converts categorical data into binary vectors.
2. **Bag of Words**: Represents text as a vector of word counts.

## One-Hot Encoding

One-hot encoding is a technique to convert categorical data into a binary vector representation, where each category is represented by a vector with a 1 in the position corresponding to that category and 0s elsewhere.

### Overview

- **Step 1:** Import necessary libraries.
- **Step 2:** Prepare sample text data.
- **Step 3:** Create tokens from the text.
- **Step 4:** Fit the one-hot encoder on the tokens.
- **Step 5:** Encode sentences using the fitted encoder.

```python
# Step 1: Import Required Libraries
from sklearn.preprocessing import OneHotEncoder
```

```python
# Step 2: Prepare Sample Text Data
# Define a list of sample documents
documents = [
    "Hi Hello how are you?",
    "Hi how are you doing?"
]

# Split each document into words
words = [doc.split() for doc in documents]

print("Documents:", documents)
print("Words per document:", words)
```

```python
# Step 3: Create Tokens
# Flatten the list of words into a list of individual tokens for fitting the encoder
tokens = [[word] for doc in words for word in doc]

print("Tokens:", tokens)
```

```python
# Step 4: Fit the One-Hot Encoder
# Initialize the encoder with sparse_output=False for dense arrays
one_hot_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder on the tokens to learn the vocabulary
one_hot_encoder.fit(tokens)

print("Vocabulary categories:", one_hot_encoder.categories_)
```

```python
# Step 5: Encode the Sentences
# For each sentence, encode each word using the fitted encoder
for sentence in words:
    print(f"Sentence: {sentence}")
    encoded_words = one_hot_encoder.transform([[word] for word in sentence])
    print("Encoded words:")
    for i, (word, encoding) in enumerate(zip(sentence, encoded_words)):
        print(f"  {word}: {encoding}")
    print()
```

## Conclusion

This notebook demonstrated two fundamental text encoding techniques:

- **One-Hot Encoding**: Useful for categorical data with a small number of categories, creating binary vectors for each word.
- **Bag of Words**: Suitable for larger text corpora, representing documents as vectors of word frequencies.

Both methods are foundational in NLP, though more advanced techniques like TF-IDF or embeddings are often preferred for complex tasks.

## Bag of Words

Bag of Words (BoW) is a text representation technique that converts text into a vector of word counts, ignoring grammar and word order but capturing the frequency of words.

### Steps

- Prepare documents.
- Use CountVectorizer to fit and transform the text.

```python
# Prepare sample documents for Bag of Words
documents = [
    "Hi Hello how are you?",
    "Hi how are you doing, Hi?"
]
```

```python
# Import CountVectorizer and fit on documents
from sklearn.feature_extraction.text import CountVectorizer

bow = CountVectorizer()

bow_matrix = bow.fit(documents)

print("Vocabulary:", bow.vocabulary_)
```

```python
# Transform documents to Bag of Words matrix
encoded_matrix = bow.transform(documents).toarray()
print("Encoded Matrix:\n", encoded_matrix)
```
