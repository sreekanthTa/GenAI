# One-Hot Encoding for Text Data

This notebook demonstrates how to perform one-hot encoding on text data using scikit-learn's `OneHotEncoder`. One-hot encoding is a technique to convert categorical data into a binary vector representation, where each category is represented by a vector with a 1 in the position corresponding to that category and 0s elsewhere.

## Overview

- **Step 1:** Import necessary libraries.
- **Step 2:** Prepare sample text data.
- **Step 3:** Create tokens from the text.
- **Step 4:** Fit the one-hot encoder on the tokens.
- **Step 5:** Encode sentences using the fitted encoder.

```python
# Step 1: Import Required Libraries
from sklearn.preprocessing import OneHotEncoder
```

```python
# Step 2: Prepare Sample Text Data
# Define a list of sample documents
documents = [
    "Hi Hello how are you?",
    "Hi how are you doing?"
]

# Split each document into words
words = [doc.split() for doc in documents]

print("Documents:", documents)
print("Words per document:", words)
```

```python
# Step 3: Create Tokens
# Flatten the list of words into a list of individual tokens for fitting the encoder
tokens = [[word] for doc in words for word in doc]

print("Tokens:", tokens)
```

```python
# Step 4: Fit the One-Hot Encoder
# Initialize the encoder with sparse_output=False for dense arrays
one_hot_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder on the tokens to learn the vocabulary
one_hot_encoder.fit(tokens)

print("Vocabulary categories:", one_hot_encoder.categories_)
```

```python
# Step 5: Encode the Sentences
# For each sentence, encode each word using the fitted encoder
for sentence in words:
    print(f"Sentence: {sentence}")
    encoded_words = one_hot_encoder.transform([[word] for word in sentence])
    print("Encoded words:")
    for i, (word, encoding) in enumerate(zip(sentence, encoded_words)):
        print(f"  {word}: {encoding}")
    print()
```

## Conclusion

This notebook showed how to apply one-hot encoding to text data. Each word is converted into a binary vector based on the learned vocabulary. This technique is useful for preparing categorical data for machine learning models, though for large vocabularies, more efficient methods like embeddings are preferred.
