```python
from nltk.tokenize import sent_tokenize

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,WordNetLemmatizer
from nltk import pos_tag

```

```python
def do_stop_words_removal(text):
    stop_words = set(stopwords.words('english'))
    removed_stop_words = [word for word in text.split() if word not in stop_words]

    return " ".join(removed_stop_words)

def do_sent_tokenize(para):
    sentences = sent_tokenize(para)
    return sentences

def do_word_tokenize(sentence):
    words = word_tokenize(sentence)
    return words

def do_stemming(words):
    ps = PorterStemmer()
    stemmed_words = []
    for word in words:
        stemmed_word = ps.stem(word)
        stemmed_words.append(stemmed_word)
    
    return " ".join(stemmed_words)
 
def do_lemmatization(words):
    lemmatizer = WordNetLemmatizer()
    lemmatized_words = []
    tagged_word = pos_tag(words)

    for tag_word in tagged_word:
        #pos tagging
       

        if tag_word[1].startswith('NN'):
            pos = 'n'
        elif tag_word[1].startswith('VB'):
            pos = 'v'
        elif tag_word[1].startswith('JJ'):
            pos = 'a'
        else:
            pos = 'n'   


        lemmatized_word = lemmatizer.lemmatize(tag_word[0], pos=pos)
        lemmatized_words.append(lemmatized_word)
    
    return " ".join(lemmatized_words)


    

```

```python
up_to_lamitization = []

for para in data: 

    #Convert to LowerCase
    para = para.lower()

    #Remove Stop Words
    removed_stop_words_text = do_stop_words_removal(para)
 
    # print("After Stop Words Removal: ", removed_stop_words_text)

    #Sentence Tokenization
    tokenized_sentencses = do_sent_tokenize(removed_stop_words_text)
    # print("After Sentence Tokenization: ", tokenized_sentencses)


    multiple_sentences = []
    
    #Word Tokenization
    for index,sentence in enumerate(tokenized_sentencses):
        # print("processing...", sentence)

 
        tokenized_words = do_word_tokenize(sentence)
        # print("After Word Tokenization: ", tokenized_words)

        #Stemming
        stemmed_text = do_stemming(tokenized_words)
        # print("After Stemming: ", stemmed_text)

        #Lemmatization
        splitted_stemmed_text = stemmed_text.split()
        # print("Splitted Stemmed Text: ", splitted_stemmed_text)
        lemmatized_text = do_lemmatization(splitted_stemmed_text)
        # print("After Lemmatization: ", lemmatized_text)

        multiple_sentences.append(lemmatized_text)
       
        # print("Before Preprocessing: ", sentence)
        tokenized_sentencses[index] = " ".join(multiple_sentences)
        # print("After Preprocessing: ", tokenized_sentencses)
    up_to_lamitization.append(tokenized_sentencses[index])
    # print("After Preprocessing: ", tokenized_sentencses[index])
        

    
print("After Preprocessing: ", up_to_lamitization)


    
    
```
