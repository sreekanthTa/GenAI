```python
import pandas as pd
```

```python
data = pd.read_csv("../phone_reviews.csv")
data.head()
```

```python
lowered_rows = [row["review"].lower() for _,row in data.iterrows()]
    
print(lowered_rows[:5])
```

Tokenization


```python
import nltk
```

```python
from nltk.tokenize import sent_tokenize

sentences = []

for para in lowered_rows:
    sentences.append(sent_tokenize(para))

print(sentences[:5])
```

```python
# Punctuation Removal
import string

punctuation_removed = []

for sentence_list in sentences:
    cleaned_sentences = []
    for sent in sentence_list:
        # Remove punctuation from each sentence
        cleaned_sent = ''.join(char for char in sent if char not in string.punctuation)
        cleaned_sentences.append(cleaned_sent)
    punctuation_removed.append(cleaned_sentences)

# Update sentences with cleaned version
sentences = punctuation_removed

print(sentences[:5])
```

```python
# Stemming
from nltk.stem import RegexpStemmer
stemmer = RegexpStemmer("!$")

for index1, sentence in enumerate(sentences):
    for index2, para in enumerate(sentence):
        words = [stemmer.stem(word) for word in para.split()]
        joined_ = " ".join(words)
        print(joined_)
        sentences[index1][index2] = joined_

```

```python
# Stemming
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

for index1, sentence in enumerate(sentences):
    for index2, para in enumerate(sentence):
        words = [stemmer.stem(word) for word in para.split()]
        joined_ = " ".join(words)
        print(joined_)
        sentences[index1][index2] = joined_

```

```python
# Stemming
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer("english")

for index1, sentence in enumerate(sentences):
    for index2, para in enumerate(sentence):
        words = [stemmer.stem(word) for word in para.split()]
        joined_ = " ".join(words)
        sentences[index1][index2] = joined_

print(sentences[:5])
```

```python
# Lamitization
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

for index1, sentence in  enumerate(sentences):
    for index2, para in enumerate(sentence):
        words = nltk.word_tokenize(para)
        pos_tags = nltk.pos_tag(words)
        lamitized= []
        for word, tag in pos_tags:
            if tag.startswith('NN'):
                pos = 'n'
            elif tag.startswith('VB'):
                pos = 'v'
            elif tag.startswith('JJ'):
                pos = 'a'
            elif tag.startswith('RB'):
                pos= 'r'
            else:
                pos = 'n'
            lamitized.append(lemmatizer.lemmatize(word, pos))
        sentences[index1][index2] = " ".join(lamitized)
            
        # print(pos_tags)
print(sentences)
```

```python
# Stop Words 
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
without_stop_words_se = []

for index, sentence in enumerate(sentences):
   for sent in sentence:
      without_stop_words = []
      for word in sent.split():
         if word not in set(stopwords.words("english")):
            without_stop_words.append(word)
      without_stop_words_sentence = " ".join(without_stop_words)
      without_stop_words_se.append(without_stop_words_sentence)
   
print(without_stop_words_se)
```

ONE HOT ENCODER

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

one_hot_encoder = OneHotEncoder(sparse_output= False)

one_array = []
for index, sentence in enumerate(without_stop_words_se):
    words = np.array(sentence.split()).reshape(-1, 1)

    encoded_sentence = one_hot_encoder.fit_transform(words)
    one_array.append(encoded_sentence)



print(one_array)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer()

X = count_vectorizer.fit_transform(without_stop_words_se)

print(X.toarray())
print(count_vectorizer.get_feature_names_out())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()

X = tfidf_vectorizer.fit_transform(without_stop_words_se)

print(X.toarray())
print(tfidf_vectorizer.get_feature_names_out())
```
